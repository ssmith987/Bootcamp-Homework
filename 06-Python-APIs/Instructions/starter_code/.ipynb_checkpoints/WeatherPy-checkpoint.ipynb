### Analysis
* As expected, the weather becomes significantly warmer as one approaches the equator (0 Deg. Latitude). More interestingly, however, is the fact that the southern hemisphere tends to be warmer this time of year than the northern hemisphere. This may be due to the tilt of the earth.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print(len(cities))
cities
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
column_names = ["City", "Cloudiness", "Country", "Date", "Humidity", "Lat", "Lng", "Max Temp", "Wind Speed"]
weather_data_df = pd.DataFrame(columns=column_names)
weather_data_df['City'] = cities

weather_data_df.head()
```

```python
record = 0
set_number = 1

print('Beginning Data Retrieval')
print('-----------------------------')

start_time = time.time()

for index,row in weather_data_df.iterrows():
    record += 1
    
    try:
        print(f"Processing Record {record} of Set {set_number} | {row['City']}")
        url = "http://api.openweathermap.org/data/2.5/weather?q="
        query_url = url + row['City'] + "&APPID=" + api_key
        weather_request = requests.get(query_url).json()
        
        row['Cloudiness'] = weather_request['clouds']['all']
        row['Country'] = weather_request['sys']['country']
        row['Date'] = weather_request['dt']
        row['Humidity'] = weather_request['main']['humidity']
        row['Lat'] = weather_request['coord']['lat']
        row['Lng'] = weather_request['coord']['lon']
        row['Max Temp'] = weather_request['main']['temp_max']
        row['Wind Speed'] = weather_request['wind']['speed']
            
    except:
        print("City not found.  Skipping...")
        continue
            
    if (index + 1) % 60 == 0:
        run_time = time.time() - start_time
        time.sleep(60 - run_time)
        start_time = time.time()
        set_number = set_number + 1
        record = 0
print("""-----------------------------
Data Retrieval Complete
-----------------------------""")

```

```python
weather_data_df.dropna(axis=0, how='any', inplace=True)
weather_data_df.reset_index(drop=True)
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.

```python
weather_data_df.to_csv("weather_data.csv", encoding="utf-8", index=False)
weather_data_df.count()
```
