```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
city_ride_data = pd.merge(ride_data,city_data,on="city", how="left")
```

```python
# Display the data table for preview
city_ride_data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Group data by type and city with aggregate calculations
city_group_data = city_ride_data.groupby(["type","city"]).agg({'ride_id':'count', 'fare':'mean',
                                                                              'driver_count':'max'})
city_group_data.head()
```

```python
# Obtain the x and y coordinates for each of the three city types
rural_x = city_group_data.loc['Rural','ride_id']
rural_y = city_group_data.loc['Rural','fare']
rural_driver = city_group_data.loc['Rural','driver_count']

suburban_x = city_group_data.loc['Suburban','ride_id']
suburban_y = city_group_data.loc['Suburban','fare']
suburban_driver = city_group_data.loc['Suburban','driver_count']

urban_x = city_group_data.loc['Urban','ride_id']
urban_y = city_group_data.loc['Urban','fare']
urban_driver = city_group_data.loc['Urban','driver_count']
```

```python
# Build the scatter plots for each city types
rural = plt.scatter(rural_x, rural_y, s=rural_driver*10, c='gold', alpha=.75, edgecolors='black')
suburban = plt.scatter(suburban_x, suburban_y, s=suburban_driver*10, c='lightskyblue', alpha=.75, edgecolors='black')
urban = plt.scatter(urban_x, urban_y, s=urban_driver*10, c='lightcoral', alpha=.75, edgecolors='black')

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides(Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()


# Create a legend
legend = plt.legend((rural, suburban, urban),("Rural", "Suburban", "Urban"), loc=0, title="City Types")
legend.legendHandles[0]._sizes = [40]
legend.legendHandles[1]._sizes = [40]
legend.legendHandles[2]._sizes = [40]

# Incorporate a text label regarding circle size
plt.text(45,40,"Note: Circle size correlates with driver count per city.")

# Save Figure
plt.savefig("pyber_data.png")

# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
city_type_fares = city_ride_data.groupby(['type'], as_index=False).sum()

# Build Pie Chart
plt.pie(city_type_fares['fare'], labels=city_type_fares['type'], colors=['gold','lightskyblue','lightcoral'], 
        explode=(0,0,.1), autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")

# Save Figure
plt.savefig("fares_by_city_type.png")

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Type Percents
city_type_rides = city_ride_data.groupby(['type'], as_index=False).count()

# Build Pie Chart
plt.pie(city_type_rides['ride_id'], labels=city_type_rides['type'], colors=['gold','lightskyblue','lightcoral'], 
        explode=(0,0,.1), autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Rides by City Type")

# Save Figure
plt.savefig("rides_by_city_type.png")

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Type Percents
city_type_drivers = city_ride_data.groupby(['type', 'city'], as_index=False).max()
type_drivers = city_type_drivers.groupby(['type'],as_index=False).sum()

# Build Pie Chart
plt.pie(type_drivers['driver_count'], labels=type_drivers['type'], colors=['gold','lightskyblue','lightcoral'], 
        explode=(0,0,.1), autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Drivers by City Type")

# Save Figure
plt.savefig("drivers_by_city_type.png")

# Show Figure
plt.show()
```
